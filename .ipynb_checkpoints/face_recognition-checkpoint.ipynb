```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
```

```python
#still from original video
base = cv2.imread('still_from_original.png', 1)
base_rgb = cv2.cvtColor(base, cv2.COLOR_BGR2RGB)
plt.imshow(base_rgb)
```

```python
cap = cv2.VideoCapture('./samples/megamind.avi')
```

```python
video_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
video_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
video_width, video_height = int(video_width), int(video_height)
```

```python
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('face_and_eyes_detection_result.avi', fourcc, 30.0, (video_width,video_height))
```

```python
face_cascade = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./cascades/haarcascade_eye.xml')
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #faces detection
    faces = face_cascade.detectMultiScale(gray, 1.2, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255,0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_face = frame[y:y+h, x:x+w]
        
        #eyes detection
        eyes = eye_cascade.detectMultiScale(roi_face, 1.2, 12)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_face, (ex,ey), (ex+ew, ey+eh), (0,0,255), 2)
    
    out.write(frame)
    cv2.imshow('facial rec', frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
cap.release()
```
